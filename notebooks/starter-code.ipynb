### Idenfity Issues

Feel free to break this up across multiple cells

````python
# TODO: Check for missing values
print("=== Missing Values ===")
print(sales.isnull().sum())

# TODO: Check for duplicates
print("\n=== Duplicates ===")
print("Duplicate rows:", sales.duplicated().sum())
print("Duplicate order IDs:", sales["order_id"].duplicated().sum())
# TODO: Check for invalid revenue
print("\n=== Invalid Revenue ===")
sales["revenue_clean"] = pd.to_numeric(
    sales["revenue"].astype(str).str.replace(r"[$,]", "", regex=True), 
    errors="coerce")

print("Negative revenue:", (sales["revenue_clean"] < 0).sum())
print("Zero revenue:", (sales["revenue_clean"] == 0).sum())
print("Null revenue after cleaning:", sales["revenue_clean"].isnull().sum())

# TODO: Check for invalid quantities
print("\n=== Invalid Quantities ===")
print("Zero quantity:", (sales["quantity"] == 0).sum())
print("Negative quantity:", (sales["quantity"] < 0).sum())
# TODO: Check for inconsistent categories
print("\n=== Inconsistent Categories ===")
print("Unique status values:")
print(sales["status"].value_counts())
print("\nUnique payment methods:")
print(sales["payment_method "].value_counts())
# TODO: Check for invalid dates
print("\n=== Invalid Dates ===")
sales["order_date"] = pd.to_datetime(sales["order_date"], errors="coerce")
print("Future dates:", (sales["order_date"] > pd.Timestamp.today()).sum())
print("Null dates:", sales["order_date"].isnull().sum())
````

---

### Create Validation Columns

````python
# TODO: Create validation flags

# Example:
# sales["missing_customer_id"] = ...
sales["missing_customer_id"] = sales["customer_id"].isnull()
sales["negative_revenue"] = sales["revenue_clean"] < 0
sales["zero_quantity"] = sales["quantity"] == 0
sales["future_date"] = sales["order_date"] > pd.Timestamp.today()
sales["duplicate_order_id"] = sales["order_id"].duplicated(keep=False)
sales["invalid_unit_price"] = sales["unit_price"] <= 0
sales["null_revenue"] = sales["revenue_clean"].isnull()
sales.head()
````

---

### Summarize Issues



````python
# TODO: Summarize validation columns
flag_cols = [
    "missing_customer_id",
    "negative_revenue",
    "zero_quantity",
    "future_date",
    "duplicate_order_id",
    "invalid_unit_price",
    "null_revenue"
]
print("=== Validation Summary ===")
summary = pd.DataFrame({
    "flag": flag_cols,
    "issue_count": [sales[flag].sum() for flag in flag_cols],
    "percentage": [f"{(sales[flag].sum() / len(sales) * 100):.1f}%" for flag in flag_cols]
})
display(summary)
````

---

### Assertions

````python
# TODO: Write assertions
print("=== Running Assertions ===")

try:
    assert sales["order_id"].notnull().all(), "order_id has null values"
    print("order_id has no nulls")
except AssertionError as e:
    print(f"FAIL: {e}")

try:
    assert (sales["quantity"] > 0).all(), "quantity has zero or negative values"
    print("quantity all positive")
except AssertionError as e:
    print(f"FAIL: {e}")
try:
    assert (sales["unit_price"] > 0).all(), "unit_price has zero or negative values"
    print("unit_price all positive")
except AssertionError as e:
    print(f"FAIL: {e}")

try:
    assert sales["revenue_clean"].notnull().all(), "revenue has null or unparseable values"
    print("revenue has no nulls")
except AssertionError as e:
    print(f"FAIL: {e}")

try:
    assert (sales["revenue_clean"] >= 0).all(), "revenue has negative values"
    print("revenue all non-negative")
except AssertionError as e:
    print(f"FAIL: {e}")

try:
    assert (sales["order_date"] <= pd.Timestamp.today()).all(), "order_date has future dates"
    print("order_date no future dates")
except AssertionError as e:
    print(f"FAIL: {e}")

try:
    assert not sales.duplicated().any(), "dataset has duplicate rows"
    print("no duplicate rows")
except AssertionError as e:
    print(f"FAIL: {e}")

print("\n=== Assertions Complete ===")
````

---

### Now clean your dataset

Save it as `clean_sales`

````python
# TODO: Filter out problematic rows
sales["has_any_issue"] = sales[flag_cols].any(axis=1)
clean_sales = sales[~sales["has_any_issue"]].copy()
clean_sales = clean_sales.reset_index(drop=True)
````

Reference the lab `README` to remember which metrics to investigate.
